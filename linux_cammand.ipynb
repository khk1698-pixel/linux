#ls 명령어 설명
ls (list segments) 명령어는 리눅스/유닉스 환경에서 디렉토리 내부의 파일과 하위 디렉토리 목록을 화면에 출력하는 핵심 명령어입니다. 기본적으로 현재 디렉토리의 내용을 보여주며, 다양한 옵션을 통해 상세 정보, 숨김 파일, 정렬 방식 등을 제어할 수 있습니다.

#주요 ls 명령어 옵션
ls -l: 상세 정보(권한, 소유자, 크기, 수정 시간)를 롱 포맷으로 표시.
ls -a: 숨김 파일(.)을 포함한 모든 파일 표시.
ls -al (또는 ls -la): 숨김 파일을 포함하여 모든 파일의 상세 정보를 표시 (가장 많이 사용).
ls -lh: 파일 크기를 읽기 쉬운 단위()로 표시.
ls -t: 파일의 수정 시간 순서대로 정렬.
ls -r: 목록을 역순으로 표시.
ls -R: 하위 디렉토리의 내용까지 재귀적으로 표시.

#cd 명령어 설명
cd (Change Directory) 명령어는 리눅스, 유닉스, 윈도우 등 운영체제 터미널에서 현재 작업 디렉토리(폴더)를 이동하는 가장 기본적인 명령어입니다. 주로 cd [이동할 디렉토리 경로] 형태로 사용하며, 절대 경로와 상대 경로를 모두 지원하여 편리하게 파일 시스템을 탐색할 수 있습니다.
#핵심 cd 명령어 및 팁
cd: 홈 디렉토리($HOME)로 이동
cd ~: 홈 디렉토리로 이동
cd ..: 상위 디렉토리(부모 폴더)로 이동
cd -: 바로 이전 작업 디렉토리로 이동
cd /: 시스템의 최상위 루트 디렉토리로 이동
cd /path/to/dir: 절대 경로로 직접 이동
cd dir1/dir2: 상대 경로(현재 위치 기준)로 이동

#pwd 명령어 설명
pwd (Print Working Directory)는 리눅스 및 유닉스 계열 운영체제에서 현재 사용자가 작업 중인 디렉토리의 절대 경로를 터미널 화면에 출력하는 명령어입니다. 루트 디렉토리(/)부터 시작하는 전체 경로를 보여주어 현재 위치를 확인할 때 사용합니다.

#핵심 특징 및 사용법
용도: 파일 작업 중 위치를 잃어버리거나, 스크립트 작성 시 현재 경로를 확인해야 할 때 필수적입니다.
사용법: 터미널에 pwd 입력 후 Enter.
출력: /home/user/documents와 같은 절대 경로 형태.
옵션: pwd -P는 symbolic link 대신 실제 물리적 경로를 표시합니다.

#cp 명령어
cp (copy) 명령어는 리눅스/유닉스 환경에서 파일이나 디렉토리를 복사하는 데 사용됩니다. 원본 파일은 그대로 유지하며, 이름이나 위치를 바꾸어 새로운 복사본을 생성할 수 있습니다. 기본 구조는 cp [옵션] <원본> <대상>입니다.
# cp 주요 활용법 및 옵션
파일 복사: cp file1.txt file1_backup.txt (file1.txt를 file1_backup.txt로 복사)
디렉토리 복사: cp -r dir1 dir2 (디렉토리 및 하위 내용까지 모두 복사)
복사 시 묻지 않고 덮어쓰기: \cp -f source target (alias 설정 무시하고 강제 덮어쓰기)
파일 속성 유지 복사: cp -p source target (원본 파일의 권한, 시간 등 정보 유지)
복사 과정 확인: cp -v source target (복사 진행 상황 화면 출력)

#rm 명령어
rm (remove) 명령어는 유닉스 및 리눅스 계열 시스템에서 파일이나 디렉터리를 삭제하는 데 사용됩니다. 주요 옵션으로 -r (하위 디렉터리 포함 삭제)과 -f (강제 삭제)가 있으며, 한 번 삭제한 파일은 복구가 어려우므로 사용 시 주의가 필요합니다.
#rm 주요 사용법 및 옵션
파일 삭제: rm [파일명] (예: rm test.txt)
파일 여러 개 삭제: rm [파일1] [파일2]
디렉터리 및 하위 내용 삭제: rm -r [디렉터리명]
강제 삭제 (확인 메시지 없음): rm -f [파일명]
디렉터리 강제 삭제: rm -rf [디렉터리명]
대화형 삭제 (확인 후 삭제): rm -i [파일명]

#mv 명령어
mv (move) 명령어는 리눅스 및 유닉스 시스템에서 파일이나 디렉토리를 다른 위치로 이동하거나, 이름을 변경할 때 사용하는 핵심 명령어입니다. 기본적으로 mv [옵션] [원본] [대상] 형식을 사용하며, 원본 파일은 삭제되고 대상 위치에 새로운 파일로 저장됩니다.
#핵심 용도 및 사용법
파일/디렉토리 이동: mv file1.txt /home/user/documents/ (file1.txt를 documents 폴더로 이동).
파일/디렉토리 이름 변경: mv old_name.txt new_name.txt (파일 이름을 old에서 new로 변경).
여러 파일 이동: mv file1.txt file2.txt /folder/ (여러 파일을 한 번에 폴더로 이동).
