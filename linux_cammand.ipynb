#ls 명령어 설명
ls (list segments) 명령어는 리눅스/유닉스 환경에서 디렉토리 내부의 파일과 하위 디렉토리 목록을 화면에 출력하는 핵심 명령어입니다. 기본적으로 현재 디렉토리의 내용을 보여주며, 다양한 옵션을 통해 상세 정보, 숨김 파일, 정렬 방식 등을 제어할 수 있습니다.

#주요 ls 명령어 옵션
ls -l: 상세 정보(권한, 소유자, 크기, 수정 시간)를 롱 포맷으로 표시.
ls -a: 숨김 파일(.)을 포함한 모든 파일 표시.
ls -al (또는 ls -la): 숨김 파일을 포함하여 모든 파일의 상세 정보를 표시 (가장 많이 사용).
ls -lh: 파일 크기를 읽기 쉬운 단위()로 표시.
ls -t: 파일의 수정 시간 순서대로 정렬.
ls -r: 목록을 역순으로 표시.
ls -R: 하위 디렉토리의 내용까지 재귀적으로 표시.

#cd 명령어 설명
cd (Change Directory) 명령어는 리눅스, 유닉스, 윈도우 등 운영체제 터미널에서 현재 작업 디렉토리(폴더)를 이동하는 가장 기본적인 명령어입니다. 주로 cd [이동할 디렉토리 경로] 형태로 사용하며, 절대 경로와 상대 경로를 모두 지원하여 편리하게 파일 시스템을 탐색할 수 있습니다.

cd /path/to/dir: 절대 경로로 직접 이동
cd dir1/dir2: 상대 경로(현재 위치 기준)로 이동

#cp 명령어
cp (copy) 명령어는 리눅스/유닉스 환경에서 파일이나 디렉토리를 복사하는 데 사용됩니다. 원본 파일은 그대로 유지하며, 이름이나 위치를 바꾸어 새로운 복사본을 생성할 수 있습니다. 기본 구조는 cp [옵션] <원본> <대상>입니다.
# cp 주요 활용법 및 옵션
파일 복사: cp file1.txt file1_backup.txt (file1.txt를 file1_backup.txt로 복사)
디렉토리 복사: cp -r dir1 dir2 (디렉토리 및 하위 내용까지 모두 복사)
복사 시 묻지 않고 덮어쓰기: \cp -f source target (alias 설정 무시하고 강제 덮어쓰기)
파일 속성 유지 복사: cp -p source target (원본 파일의 권한, 시간 등 정보 유지)
복사 과정 확인: cp -v source target (복사 진행 상황 화면 출력)

#rm 명령어
rm (remove) 명령어는 유닉스 및 리눅스 계열 시스템에서 파일이나 디렉터리를 삭제하는 데 사용됩니다. 주요 옵션으로 -r (하위 디렉터리 포함 삭제)과 -f (강제 삭제)가 있으며, 한 번 삭제한 파일은 복구가 어려우므로 사용 시 주의가 필요합니다.
#rm 주요 사용법 및 옵션
파일 삭제: rm [파일명] (예: rm test.txt)
파일 여러 개 삭제: rm [파일1] [파일2]
디렉터리 및 하위 내용 삭제: rm -r [디렉터리명]
강제 삭제 (확인 메시지 없음): rm -f [파일명]
디렉터리 강제 삭제: rm -rf [디렉터리명]
대화형 삭제 (확인 후 삭제): rm -i [파일명]

#mv 명령어
mv (move) 명령어는 리눅스 및 유닉스 시스템에서 파일이나 디렉토리를 다른 위치로 이동하거나, 이름을 변경할 때 사용하는 핵심 명령어입니다. 기본적으로 mv [옵션] [원본] [대상] 형식을 사용하며, 원본 파일은 삭제되고 대상 위치에 새로운 파일로 저장됩니다.
#핵심 용도 및 사용법
파일/디렉토리 이동: mv file1.txt /home/user/documents/ (file1.txt를 documents 폴더로 이동).
파일/디렉토리 이름 변경: mv old_name.txt new_name.txt (파일 이름을 old에서 new로 변경).
여러 파일 이동: mv file1.txt file2.txt /folder/ (여러 파일을 한 번에 폴더로 이동).


#cat 설명
cat (concatenate) 명령어는 리눅스/유닉스 환경에서 파일 내용을 화면에 출력하거나, 파일을 합치고 생성하는 데 사용되는 가장 기초적이고 강력한 도구입니다. 파일 확인, 파일 내용 연속 출력, 리다이렉션(>)을 이용한 새 파일 생성 및 내용 추가가 주된 기능입니다.

#주요 활용법 및 예제
파일 내용 확인: cat filename (파일의 전체 내용 출력).
여러 파일 합치기: cat file1 file2 > combined.txt (file1, file2를 하나로 합쳐 combined.txt 생성).
파일 생성 (간단한 내용): cat > newfile.txt (내용 입력 후 Ctrl+D로 저장 및 종료).
줄 번호 포함 출력: cat -n filename (빈 줄 포함하여 줄 번호 표시).
파일 끝 표시: cat -E filename (각 줄의 끝에 $ 문자를 표시).
탭(Tab) 문자를 ^I로 출력: cat -T filename.

#chmod 설명
chmod(change mode)는 유닉스 및 리눅스 계열 운영체제에서 파일이나 디렉토리의 접근 권한(Permission)을 변경하는 명령어입니다. 사용자(u), 그룹(g), 기타 사용자(o)에 대해 읽기(r, 4), 쓰기(w, 2), 실행(x, 1) 권한을 8진수 숫자나 심볼릭 기호로 설정하여 보안을 관리합니다.
#주요 특징 및 사용법
기본 구조: chmod [옵션] [권한] [파일/디렉토리]
8진수 모드: 숫자의 합으로 권한 설정 (예: 755 = , 644 = )
심볼릭 모드: 문자를 사용하여 권한 수정 (예: u+x는 사용자에게 실행 권한 추가)
재귀적 변경: -R 옵션을 사용하여 디렉토리 하위 파일까지 한 번에 권한 변경 가능

#chown 설명
chown (change owner)은 리눅스 및 유닉스 시스템에서 파일이나 디렉토리의 소유자(User) 및 그룹(Group)을 변경하는 명령어입니다. 보통 관리자 권한(root)이 필요하며, -R 옵션을 통해 하위 파일 및 디렉토리의 소유권까지 일괄 변경할 수 있습니다.

#주요 명령어 구조:
chown [옵션] [소유자]:[소유그룹] [파일/디렉토리]
사용 예시:
소유자만 변경: chown user1 file.txt
소유자와 그룹 변경: chown user1:group1 file.txt
하위 디렉토리까지 일괄 변경 (Recursive): chown -R user1:group1/path/to/dir

핵심 특징:
권한: 파일 소유자를 변경하려면 root 권한이 필요합니다.
chgrp과 차이: chown은 소유자와 그룹 모두, chgrp은 그룹만 변경합니다.
-R 옵션: 디렉토리 내부의 모든 파일과 서브 디렉토리의 소유권을 변경합니다.

#ln 설명
리눅스 ln (link) 명령어는 파일이나 디렉터리에 대한 링크(바로가기)를 생성하여, 하나의 파일을 여러 이름으로 접근하게 하는 도구입니다. 기본적으로 하드 링크를 생성하며, -s 옵션을 사용하여 심볼릭 링크(바로가기)를 생성할 수 있습니다.

#주요 내용 및 특징
기본 문법: ln [옵션] [원본파일] [링크파일]
심볼릭 링크 (Symbolic Link, -s):
윈도우의 '바로가기'와 유사합니다.
원본 파일이 삭제되면 링크 파일은 동작하지 않습니다.
파일뿐만 아니라 디렉터리 링크도 가능합니다.
예: ln -s [원본] [링크]
하드 링크 (Hard Link, 기본):
원본 파일과 동일한 inode를 공유하는 또 다른 파일 이름을 생성합니다.
원본을 삭제해도 링크 파일은 남아서 작동합니다.
파일 내용이 서로 동기화됩니다.
디렉터리 링크는 불가능합니다.
예: ln [원본] [링크]

#tar 설명
tar(Tape ARchiver)는 리눅스/유닉스 환경에서 여러 파일과 디렉토리를 하나의 파일로 묶거나(아카이브), 이를 해제하는 명령어입니다. 파일 속성(권한 등)을 보존하며, gzip/bzip2와 결합하여 압축(tar.gz, tar.bz2) 용도로 가장 많이 사용됩니다.
#주요 명령어 (자주 사용되는 조합)
파일 묶기 (압축 안 함): tar -cvf [아카이브파일명.tar] [대상파일/디렉토리]
파일 풀기 (압축 안 함): tar -xvf [아카이브파일명.tar]
tar.gz로 압축하기: tar -zcvf [파일.tar.gz] [대상파일/디렉토리]
tar.gz 압축 풀기: tar -zxvf [파일.tar.gz]
tar.gz 압축을 특정 디렉토리에 풀기: tar -zxvf [파일.tar.gz] -C [경로]

#find 설명
리눅스 find 명령어는 파일 시스템에서 조건(이름, 크기, 시간, 권한 등)에 맞는 파일이나 디렉터리를 검색하는 강력한 도구입니다. 기본 구조는 find [경로] [옵션] [표현식]이며, -name (이름), -type (타입: f-파일, d-디렉터리), -size (크기) 등이 가장 많이 사용됩니다.
#핵심 옵션 및 사용법
기본 검색: find . -name "test.txt" (현재 디렉터리에서 test.txt 찾기)
파일/디렉터리 구분: find /home -type d -name "data" (/home에서 data라는 디렉터리 검색)
크기 기준: find / -size +100M (100MB 이상인 파일 검색)
시간 기준: find . -mtime -7 (최근 7일 이내에 수정된 파일)
검색 후 명령 실행: find . -name "*.log" -delete (.log 파일 찾아 삭제)
권한 검색: find . -perm 777 (모든 권한이 부여된 파일 찾기)
